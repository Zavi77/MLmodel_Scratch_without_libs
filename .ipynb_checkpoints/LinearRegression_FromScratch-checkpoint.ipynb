```python
import pandas as pd
df = pd.read_csv('train.csv')
df.head()
```

```python
x = df['x'].tolist()
y = df['y'].tolist()
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,s=5)
plt.grid()
plt.xlabel("x values")
plt.ylabel("y values")
```

```python
def gradient_descent(x,y,w,b,alpha):
    dl_dw = 0.0
    dl_db = 0.0
    N = len(x)
    for i in range(N):
        dl_dw = dl_dw + (-1)*x[i]*(y[i]-(w*x[i]+b))
        dl_db = dl_db + (-1)*(y[i]-(w*x[i]+b))
    
    w = w - (1/float(N)) * dl_dw * alpha 
    b = b - (1/float(N)) * dl_db * alpha
    return w,b
```

```python
def cost_function(x,y,w,b):
    total_error = 0.0
    N = len(x)
    for i in range(N):
        total_error += (y[i] - (w*x[i] + b))**2
        
    return total_error/(2*float(N))
```

```python
def train(x, y, w, b, alpha, n_iter):

    for i in range(n_iter):
        w,b = gradient_descent(x,y,w,b,alpha)
        
        if i % 400 ==0:
            print("iteration:", i, "cost: ", cost_function(x,y,w,b))
            
    return w,b
```

```python
def predict(x,w,b):
    return x*w +b
```

```python
w,b = train(x, y, 0, 0, 0.0001, 2000)

x_new = 50.0
y_new = predict(x_new, w, b)

print(y_new)
```
